```python
from transformers import AutoTokenizer, AutoModelForTextToWaveform
import torch, scipy
from IPython.display import Audio
import os
import time
start_time = time.time()

# Load model & tokenizer
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-vie")
model = AutoModelForTextToWaveform.from_pretrained("facebook/mms-tts-vie")

# Văn bản tiếng Việt
with open('example.txt', 'r', encoding='utf-8') as file:
    text = file.read()
inputs = tokenizer(text, return_tensors="pt")

# Sinh waveform
with torch.no_grad():
    output = model(**inputs).waveform

# Chuyển sang numpy
waveform = output.squeeze().cpu().numpy()
waveform = (waveform * 32767).astype("int16")

# Lưu file .wav


# tạo folder output nếu chưa có
os.makedirs("output", exist_ok=True)

# lưu file vào thư mục output
scipy.io.wavfile.write("output/output.wav", 
                       rate=model.config.sampling_rate, 
                       data=waveform)
print("✅ Đã lưu file output.wav")

end_time = time.time()
total_time = end_time - start_time
print(f"Tổng thời gian: {total_time:.2f} giây")
# Nghe trực tiếp trong notebook
Audio(waveform, rate=model.config.sampling_rate)
```
